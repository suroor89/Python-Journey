### Introduction to Tuples
##### Explanation:
Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

#### Creating Tuples

```python
## creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
# Creating tuples with elements
tuple_with_elements = (1, 2, 3, 4, 5)
print("Tuple:", tuple_with_elements)

# Tuple with mixed data types
mixed_tuple = (10, "hello", 3.14, True, None)
print("Mixed tuple:", mixed_tuple)

# Single element tuple (note the comma)
single_element = (42,)
print("Single element:", single_element)
print("Type:", type(single_element))
```

#### Accessing Tuple Elements

```python
# Indexing - accessing elements by position
my_tuple = (10, 20, 30, 40, 50)
print("First element (index 0):", my_tuple[0])
print("Third element (index 2):", my_tuple[2])
print("Last element (index -1):", my_tuple[-1])
print("Second last element (index -2):", my_tuple[-2])
```

```python
# Slicing - getting a subset of elements
my_tuple = (10, 20, 30, 40, 50, 60, 70)
print("Elements from index 1 to 4:", my_tuple[1:4])
print("First three elements:", my_tuple[:3])
print("Last three elements:", my_tuple[-3:])
print("Every second element:", my_tuple[::2])
print("Reversed tuple:", my_tuple[::-1])
```

#### Tuple Operations

```python
# Concatenation - combining tuples
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined = tuple1 + tuple2
print("Concatenated:", combined)

# Repetition - repeating a tuple
repeated = tuple1 * 3
print("Repeated 3 times:", repeated)

# Length - number of elements
print("Length of combined tuple:", len(combined))

# Membership - checking if element exists
print("Is 5 in combined?", 5 in combined)
print("Is 10 in combined?", 10 in combined)
```

#### Immutable Nature of Tuples
Tuples cannot be modified after creation. Any attempt to change, add, or remove elements will raise an error.

```python
my_tuple = (1, 2, 3, 4, 5)

# Trying to modify an element - this will raise an error
try:
    my_tuple[0] = 10
except TypeError as e:
    print(f"Error when modifying: {e}")

# Trying to delete an element - this will also raise an error
try:
    del my_tuple[0]
except TypeError as e:
    print(f"Error when deleting: {e}")

# But we can create a new tuple by concatenation
new_tuple = (10,) + my_tuple[1:]
print("Original tuple:", my_tuple)
print("New tuple:", new_tuple)
```

#### Tuple Methods
Tuples have only two built-in methods since they are immutable.

```python
# count() - returns the number of occurrences of an element
my_tuple = (1, 2, 2, 3, 3, 3, 4, 4, 4, 4)
print("Count of 3:", my_tuple.count(3))
print("Count of 2:", my_tuple.count(2))
print("Count of 5:", my_tuple.count(5))

# index() - returns the index of the first occurrence
print("\nIndex of first 3:", my_tuple.index(3))
print("Index of first 4:", my_tuple.index(4))

# Find index starting from a specific position
print("Index of 2 starting from position 1:", my_tuple.index(2, 1))
```

#### Packing and Unpacking Tuples
Packing: combining values into a tuple
Unpacking: extracting values from a tuple into variables

```python
# Packing - creating a tuple from values
a = 10
b = 20
c = 30
packed_tuple = (a, b, c)
print("Packed tuple:", packed_tuple)

# Unpacking - extracting values from a tuple
x, y, z = packed_tuple
print(f"Unpacked values: x={x}, y={y}, z={z}")

# Unpacking with *args to capture remaining elements
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(f"First: {first}, Middle: {middle}, Last: {last}")

# Swapping values using unpacking
a, b = b, a
print(f"After swap: a={a}, b={b}")
```

#### Nested Tuples
Tuples can contain other tuples as elements.

```python
# Nested tuples
coordinates = ((1, 2), (3, 4), (5, 6))
print("Nested tuple:", coordinates)
print("First coordinate:", coordinates[0])
print("First element of first coordinate:", coordinates[0][0])

# Matrix representation using nested tuples
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)
print("\nMatrix:")
for row in matrix:
    print(row)

print(f"\nElement at row 1, column 2: {matrix[1][2]}")
```

#### Practical Examples and Use Cases

```python
# Use case 1: Returning multiple values from a function
def get_student_info(student_id):
    # Simulating database lookup
    return ("Alice", 85, "A")

name, score, grade = get_student_info(1)
print(f"Student: {name}, Score: {score}, Grade: {grade}")

# Use case 2: Using tuples as dictionary keys (lists cannot be keys)
student_scores = {
    ("Alice", 2024): 85,
    ("Bob", 2024): 92,
    ("Charlie", 2024): 78
}
print("\nStudent scores dictionary:")
for key, value in student_scores.items():
    print(f"{key}: {value}")

# Use case 3: Iteration with index using enumerate
colors = ("red", "green", "blue", "yellow")
print("\nColors with indices:")
for index, color in enumerate(colors):
    print(f"{index}: {color}")

# Use case 4: Immutability for data integrity
config = ("localhost", 8080, "prod")
print(f"\nConfiguration (immutable): {config}")
# This configuration cannot be accidentally modified in the code
```

#### Common Errors and Best Practices

```python
# Common Error 1: Forgetting comma for single-element tuple
wrong = (5)  # This is just an integer, not a tuple!
correct = (5,)  # This is a tuple
print(f"Type of (5): {type(wrong)}")
print(f"Type of (5,): {type(correct)}")

# Common Error 2: IndexError - accessing out of range index
my_tuple = (1, 2, 3)
try:
    print(my_tuple[10])
except IndexError as e:
    print(f"Error: {e}")

# Common Error 3: TypeError - trying to modify tuple
try:
    my_tuple[0] = 100
except TypeError as e:
    print(f"Error: {e}")

# Best Practice: Use tuples when you need immutability
# Use lists when you need to modify data
CONSTANTS = (3.14159, 2.71828)  # Tuples for constants
mutable_list = [1, 2, 3]  # Lists for changing data
mutable_list.append(4)
print(f"\nConstants (tuple): {CONSTANTS}")
print(f"Mutable list: {mutable_list}")
```

#### Summary: When to Use Tuples
✅ **Use Tuples When:**
- Data should not be modified (immutability needed)
- Using data as dictionary keys
- Returning multiple values from functions
- Need slightly better performance than lists
- Protecting data from accidental changes

❌ **Use Lists Instead When:**
- You need to add, remove, or modify elements
- Working with frequently changing data
- Need list-specific methods (append, remove, sort, etc.)

```python
# Quick Reference: Tuple Operations Cheat Sheet
print("=== TUPLE OPERATIONS CHEAT SHEET ===\n")

# Create
t = (1, 2, 3, 4, 5)
print(f"Create: {t}")

# Access
print(f"Index [0]: {t[0]}")
print(f"Index [-1]: {t[-1]}")
print(f"Slice [1:3]: {t[1:3]}")

# Concatenate & Repeat
print(f"Concatenate: {(1, 2) + (3, 4)}")
print(f"Repeat: {(1, 2) * 2}")

# Length & Membership
print(f"Length: {len(t)}")
print(f"3 in tuple: {3 in t}")

# Methods
print(f"Count of 2: {(1, 2, 2, 3).count(2)}")
print(f"Index of 3: {t.index(3)}")

# Unpack
x, y, z, *rest = t
print(f"Unpacked: x={x}, y={y}, z={z}, rest={rest}")

# Immutable
print("\nTuples are IMMUTABLE - they cannot be modified after creation!")
```
