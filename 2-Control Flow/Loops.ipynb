```python
## strings

str="Bhargav Ravalji"

for i in str:
    print(i)
```

```python
## while loop
## The while loop continues to execute as long as the condition is True.

count=0
while count<5:
    print(count)
    count=count+1
```

```python
## Loop Control Statements
## break
## The break statement exits the loop permaturely
## break sstatement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue
## The continue statement skips the current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
## The pass statement is a null operation; it does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples- Calculate the sum of first N natural numbers using a while and for loop
## while loop  

n=10   
sum=0
count=1

while count<=n:
    sum=sum+count
    count+=1

print("Sum of first 10 natural number:",sum)
```
